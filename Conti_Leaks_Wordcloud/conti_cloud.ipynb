### Read Conti Chats

```python
import os
import nltk
path = os.path.join(os.getcwd(),"translated_logs.json")
with open(path,"r") as fh:
        file_data = fh.read()
        
file_data[0:200]
```

### Setup Stopwords and Generate Wordcloud

```python
from wordcloud import WordCloud,STOPWORDS
#create stop words 
nltk.download("stopwords")
from nltk.corpus import stopwords

#list of stopwords
stopwords = stopwords.words("english")
#call the wordcloud Constructor 
WC = WordCloud(stopwords=stopwords,max_words=25,background_color="white").generate(file_data)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["figure.figsize"] = (20,5)
plt.imshow(WC)
plt.axis("off")
plt.show()
```

### Add Additional Stopwords to Improve Wordcloud

```python
stopwords.extend(["Message","user3","user4","user7","user8", "user9", "Date", "channel", "tl1", "tl2", "yes", "Hobbes", "loc", "work", "session", "mediaeveryone", "com", "http", "url", "msg", "ok", "https", "general", "well", "I'm", "one", "everything", "everyone", "tomorrow", "today", "already", "give", "like", "hello", "user", "Default Share", "onion", "body", "q3mcco35auwcstmt", "defender", "bio", "mango", "stern", "veron", "driver", "error", "cannot", "ts"])
WC = WordCloud(stopwords=stopwords,max_words=70,background_color="white").generate(file_data)
plt.imshow(WC)
plt.axis("off")
plt.show()
```
